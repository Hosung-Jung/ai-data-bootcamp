```python
# ─────────────────────────────────────────────
# [도식] 120일 학습 로드맵 한눈에 보기
#   - 환경을 갖추면 이 그림을 직접 실행해보세요.
# ─────────────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(10, 3.2))
ax.axis("off")

stages = [
    ("Adapt &\nWrangling", "#4C78A8"),
    ("Collect &\nStatistics", "#54A24B"),
    ("Machine\nLearning", "#E45756"),
    ("Final\nProject", "#B279A2"),
]
x = 0.04
w = 0.20
for i, (label, color) in enumerate(stages):
    ax.add_patch(mpatches.FancyBboxPatch(
        (x, 0.35), w, 0.34, boxstyle="round,pad=0.02",
        facecolor=color, edgecolor="none", alpha=0.92))
    ax.text(x + w / 2, 0.52, label, ha="center", va="center",
            color="white", fontsize=11, fontweight="bold")
    if i < len(stages) - 1:
        ax.annotate("", xy=(x + w + 0.03, 0.52), xytext=(x + w, 0.52),
                    arrowprops=dict(arrowstyle="-|>", color="#555", lw=2))
    x += w + 0.04

ax.text(0.02, 0.85, "Day 1  (You are here)", fontsize=11, color="#333", fontweight="bold")
ax.text(0.04, 0.14, "Today: set up tools  +  make your first commit  +  plan your 120 days",
        fontsize=10, color="#666")
ax.scatter([0.05], [0.78], s=180, color="#F58518", zorder=5)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.title("AI·Data Bootcamp — 120-Day Roadmap", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.show()
```
