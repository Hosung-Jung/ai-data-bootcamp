```python
# ─────────────────────────────────────────────
# [도식] 개발 환경 4도구의 관계
# ─────────────────────────────────────────────
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(9, 4.2))
ax.axis("off")

# VS Code = 가장 바깥(주방)
ax.add_patch(mpatches.FancyBboxPatch((0.05, 0.08), 0.9, 0.84,
    boxstyle="round,pad=0.02", facecolor="#EEF3F8", edgecolor="#4C78A8", lw=2))
ax.text(0.50, 0.86, "VS Code  (the kitchen: edit · run · manage)",
        ha="center", fontsize=11, fontweight="bold", color="#2A4D69")

# Anaconda = 그 안의 큰 박스
ax.add_patch(mpatches.FancyBboxPatch((0.12, 0.16), 0.76, 0.58,
    boxstyle="round,pad=0.02", facecolor="#E6F2E6", edgecolor="#54A24B", lw=2))
ax.text(0.50, 0.68, "Anaconda  (Python + data libraries bundled)",
        ha="center", fontsize=11, fontweight="bold", color="#2E6B2E")

# myenv = 가상환경
ax.add_patch(mpatches.FancyBboxPatch((0.20, 0.24), 0.60, 0.34,
    boxstyle="round,pad=0.02", facecolor="#FFF3E0", edgecolor="#F58518", lw=2))
ax.text(0.50, 0.52, "conda env: myenv  (clean workspace)",
        ha="center", fontsize=11, fontweight="bold", color="#A65A00")

# Jupyter = 안쪽
ax.add_patch(mpatches.FancyBboxPatch((0.28, 0.29), 0.44, 0.17,
    boxstyle="round,pad=0.02", facecolor="#FCE7EA", edgecolor="#E45756", lw=2))
ax.text(0.50, 0.375, "Jupyter Notebook\n(write · run · see results)",
        ha="center", va="center", fontsize=10, fontweight="bold", color="#A8323A")

plt.title("How the four tools nest together", fontsize=12, fontweight="bold")
plt.tight_layout()
plt.show()
```
