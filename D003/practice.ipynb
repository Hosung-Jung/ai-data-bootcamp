```python
# 예제: 기본 자료형 4종을 확인합니다. type()은 값의 자료형을 알려줍니다.
age = 35              # 정수(int)
avg_amount = 19900.0  # 실수(float)
region = "서울"        # 문자열(str)
is_premium = True     # 불리언(bool)

for value in [age, avg_amount, region, is_premium]:
    print(f"{str(value):>8}  →  {type(value).__name__}")
```
