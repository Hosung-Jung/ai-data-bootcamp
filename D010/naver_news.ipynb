```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://news.naver.com/section/103"  # 생활/문화 섹션

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
                  'AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/120.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers, timeout=5)
print("상태 코드:", response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
selector = 'ul[id^="_SECTION_HEADLINE_LIST"]'

headlines = soup.select_one(selector)
for i, li in enumerate(headlines.select("li"), 1):
    title = li.select_one("strong.sa_text_strong").get_text(strip=True)
    press = li.select_one("div.sa_text_press").get_text(strip=True)
    print(f"{i}. {title} ({press})")
```
